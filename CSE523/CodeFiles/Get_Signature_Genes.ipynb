```python
#Make a python list of the marker genes - 1204 marker genes (also remove nans from the marker genes excel file)
requiredGenes = []
t = []
for i in range(len(requiredGenesList)):
    for j in range(len(requiredGenesList[i])):
        if requiredGenesList[i][j]!='nan':
            t.append(requiredGenesList[i][j])
req = []
for i in range(len(t)):
    if t[i]!='nan':
        req.append(t[i])
required = np.asarray(req)
requiredGenes = []
requiredGenes = [x for x in required if str(x) != 'nan']    
print ((requiredGenes))
```
