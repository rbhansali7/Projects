```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import loadtxt
%matplotlib inline
```

```python
#Read Bulk Data file
#singleCellDF = pd.read_csv('final_single_cell_data.txt', sep='\t',header=None)
fullBulkData = pd.read_csv('TCGA_BRCA_dat.txt', sep='\t')
```

```python
fullBulkData.head()
```

```python
fullGeneNames = list(fullBulkData.columns.values)
```

```python
#Preprocess the gene names to ignore stuff after '|'
geneNames = []
for gene in fullGeneNames:
    for i in range(len(gene)):
        if gene[i]=='|':
            pos = i
            geneNames.append(gene[0:pos])
            break
```

```python
#singleCellDF = singleCellDF[1:] #take the data less the header row
fullBulkData.columns = geneNames #set the header row as the df header
```

```python
fullBulkData.head()
```

```python
#Convert FPKM to TPM
fullBulkData = fullBulkData.apply(lambda x: x*1000000/sum(x))
```

```python
fullBulkData.head()
```

```python
#Apply log2(TPM + 1)
fullBulkData = fullBulkData.apply(lambda x: np.log2(x+1))
```

```python
fullBulkData.head()
```

```python
bulkData = fullBulkData.transpose()
```

```python
bulkData.head()
```

```python
#Save the bulk data file
bulkData.to_csv('bulk_data.txt', sep='\t')
```
