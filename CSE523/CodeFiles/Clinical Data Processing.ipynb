```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import loadtxt
%matplotlib inline
```

```python
clinicalDataDF = pd.read_csv('TCGA_clincical_BRCA.txt', sep='\t')
```

```python
clinicalDataDF.head()
```

```python
clinicalDataDF['IHC_HER2'].head()
```

```python
clinicalDataDF = clinicalDataDF.loc[clinicalDataDF['IHC_HER2'] == 'Positive']
```

```python
def removeLastHyphenOnwards(x):
    for i in range(len(x)-1,0,-1):
        if x[i]=='-':
            return x[0:i]
```

```python
clinicalDataDF['CASE_ID'] = clinicalDataDF['CASE_ID'].apply(lambda x: removeLastHyphenOnwards(x))
```
