```python
#Read single Cell file
#singleCellDF = pd.read_csv('final_single_cell_data.txt', sep='\t',header=None)
singleCellDF = pd.read_csv('temp_single_cell.txt', sep='\t')
```

```python
#Calculate the average and take log2(TPM+1)
singleCellDF['BC_01'] = np.log(singleCellDF.iloc[:,15:41].astype(float).mean(axis=1).add(1))
singleCellDF['BC_02'] = np.log(singleCellDF.iloc[:,41:97].astype(float).mean(axis=1).add(1))
singleCellDF['BC_03'] = np.log(singleCellDF.iloc[:,97:134].astype(float).mean(axis=1).add(1))
singleCellDF['BC_03LN'] = np.log(singleCellDF.iloc[:,134:189].astype(float).mean(axis=1).add(1))
singleCellDF['BC_04'] = np.log(singleCellDF.iloc[:,189:248].astype(float).mean(axis=1).add(1))
singleCellDF['BC_05'] = np.log(singleCellDF.iloc[:,248:325].astype(float).mean(axis=1).add(1))
singleCellDF['BC_06'] = np.log(singleCellDF.iloc[:,325:350].astype(float).mean(axis=1).add(1))
singleCellDF['BC_07'] = np.log(singleCellDF.iloc[:,350:401].astype(float).mean(axis=1).add(1))
singleCellDF['BC_07LN'] = np.log(singleCellDF.iloc[:,401:454].astype(float).mean(axis=1).add(1))
singleCellDF['BC_08'] = np.log(singleCellDF.iloc[:,454:477].astype(float).mean(axis=1).add(1))
singleCellDF['BC_09'] = np.log(singleCellDF.iloc[:,477:537].astype(float).mean(axis=1).add(1))
singleCellDF['BC_10'] = np.log(singleCellDF.iloc[:,537:553].astype(float).mean(axis=1).add(1))
singleCellDF['BC_11'] = np.log(singleCellDF.iloc[:,553:564].astype(float).mean(axis=1).add(1))
```

```python
#Do necessary preprocessing to get the final signatureMatrixDF
finalSingleCellDF = singleCellDF.iloc[:,564:]
finalSingleCellDF.insert(0, 'gene_name', singleCellDF.iloc[:,0])
```
